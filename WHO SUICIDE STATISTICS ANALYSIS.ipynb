```python
import pandas as pd
import pandasql as ps
import numpy as np
import matplotlib.pyplot as plt
from pysqldf import SQLDF
sqldf = SQLDF(globals())

data=pd.read_csv("C:\\Users\\SAIKRISHNA\\Desktop\\who_suicide_statistics.csv")
modifiedData = data.dropna()

modifiedData.isnull().sum()

q1 = """SELECT sex as Gender,sum(suicides_no) as countofsuicides FROM modifiedData group by sex """

i1=ps.sqldf(q1)
print(i1)

i1.plot(x='Gender',y='countofsuicides',kind='pie')
plt.show()
```

```python
q2 = """SELECT year,sum(suicides_no) as countofsuicides FROM modifiedData group by year """

i2=ps.sqldf(q2)
print(i2)
i2.plot(x='year',y='countofsuicides',kind='boxplot')
plt.show()
```
